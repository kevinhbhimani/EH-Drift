```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>")) #makes the notebook wider
plt.style.use('default')
```

```python
directory_20m = '/Users/kevinhbhimani/Desktop/GPU_work/siggen_ccd/waveforms/5000.00_keV/grid_0.0200/self_repulsion_0/P42575A/q=0.00/'
directory_10m = '/Users/kevinhbhimani/Desktop/GPU_work/siggen_ccd/waveforms/5000.00_keV/grid_0.0100/self_repulsion_0/P42575A/q=0.00/'

imp_rad=15.0
imp_height=5.0
time_step=0.20
time=np.arange(0,800,time_step)
```

```python
data_e_10m_bulk=np.loadtxt(directory_10m+ 'signal_r={:.2f}_phi=0.00_z={:.2f}.txt'.format(imp_rad, imp_height))
data_e_20m_bulk=np.loadtxt(directory_20m+ 'signal_r={:.2f}_phi=0.00_z={:.2f}.txt'.format(imp_rad, imp_height))
data_e_10m_sf=np.loadtxt(directory_10m+ 'signal_r={:.2f}_phi=0.00_z={:.2f}.txt'.format(imp_rad, 0.02))
data_e_20m_sf=np.loadtxt(directory_20m+ 'signal_r={:.2f}_phi=0.00_z={:.2f}.txt'.format(imp_rad, 0.02))
```

```python
plt.plot(time,data_e_10m_bulk,'-',markersize=2,color='tab:purple')
plt.plot(time,data_e_20m_bulk,'-',markersize=2,color='tab:green')

plt.legend(['10 micron', '20 micron'])

plt.xlabel('Time (ns)')
plt.ylabel('Fraction of charge carriers collected')
plt.title('Singal at $r=15.00,z=5.00$, diffusion=0, Energy=$5$MeV')
```

```python
plt.plot(time,data_e_10m_sf,'-',markersize=2,color='tab:purple')
plt.plot(time,data_e_20m_sf,'-',markersize=2,color='tab:green')

plt.legend(['10 micron', '20 micron'])

plt.xlabel('Time (ns)')
plt.ylabel('Fraction of charge carriers collected')
plt.title('Singal at $r=15,z=0.02$, diffusion=0, Energy=$5$MeV')
```
