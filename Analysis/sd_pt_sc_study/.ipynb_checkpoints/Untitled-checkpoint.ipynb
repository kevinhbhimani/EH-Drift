```python
import os
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Directory containing the waveform files
directory = '/nas/longleaf/home/kbhimani/siggen_ccd/waveforms/P42575A/sc_pt_sc_study/'

# Initialize an empty list to store data
waveforms_data = []

# Iterate over files in the directory
for filename in tqdm(os.listdir(directory)):
    if filename.endswith('.h5') or filename.endswith('.hdf5'):
        with h5py.File(os.path.join(directory, filename), 'r') as file:
            # Check for 'event_data' dataset in the file
            event_data = file['event_data']
            waveforms = event_data['waveform']
            # Extract file-level attributes
            grid = file.attrs['grid']
            passivated_thickness = file.attrs['passivated_thickness']
            surface_drift_factor = file.attrs['surface_bulk_vel_factor']  # Assuming this attribute is available
            detector_name_bytes = file.attrs['detector_name'][:]
            detector_name = detector_name_bytes.tobytes().decode('utf-8')            
            # Extract parameters for each event
            for i in range(event_data.shape[0]):
                eng, r, z, surface_charge = event_data[i]['energy'], event_data[i]['radius'], event_data[i]['height'], event_data[i]['surface_charge']

                waveform = waveforms[i]

                # Append to the list as a dictionary
                waveforms_data.append({
                    'r': r, 
                    'z': z, 
                    'eng': eng, 
                    'sc': surface_charge,
                    'sf_drift': surface_drift_factor,
                    'pass_thickness': passivated_thickness, 
                    'grid': grid, 
                    'det': detector_name, 
                    'wf': waveform
                })
                
# Convert the list of dictionaries to a DataFrame
waveforms_df = pd.DataFrame(waveforms_data)

waveforms_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

def plot_waveforms_variation_with_colormap(df, varying_param, fixed_params, title, tolerance=1e-4):
    """
    df: DataFrame containing waveform data and parameters.
    varying_param: The parameter name that varies in the plot.
    fixed_params: Dictionary of parameters that are fixed with their values.
    title: Title for the plot.
    tolerance: Numerical tolerance for querying floating-point values.
    """
    plt.figure(figsize=(12, 8))

    # Filter dataframe based on fixed parameters
    query = " & ".join([f"({param} >= {value - tolerance} & {param} <= {value + tolerance})" if param != varying_param else "" for param, value in fixed_params.items()])
    filtered_df = df.query(query)

    # Get unique values of the varying parameter
    unique_values = sorted(filtered_df[varying_param].unique())
    cmap = cm.get_cmap('jet', len(unique_values))  # Get a colormap

    for i, value in enumerate(unique_values):
        # Filter for each unique value of the varying parameter
        specific_df = filtered_df[np.abs(filtered_df[varying_param] - value) <= tolerance]
        mean_wf = np.mean(np.vstack(specific_df['wf'].values), axis=0)  # Calculate mean waveform
        
        # Plot with color
        plt.plot(mean_wf, color=cmap(i), label=f"{varying_param} = {value:.4f}")

    # plt.colorbar(cm.ScalarMappable(cmap=cmap), label=varying_param)
    plt.xlabel('Time Step')
    plt.ylabel('Signal Amplitude')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# Define the fixed parameters for each scenario
fixed_params_sc = {'r': 15.00, 'z': 0.02, 'eng': 5000, 'sf_drift': 0.01, 'pass_thickness': 0.002}
fixed_params_sf_drift = {'r': 15.00, 'z': 0.02, 'eng': 5000, 'sc': -0.30, 'pass_thickness': 0.002}
fixed_params_pass_thickness = {'r': 15.00, 'z': 0.02, 'eng': 5000, 'sc': -0.30, 'sf_drift': 0.01}

# Plot for each scenario with a tolerance in the query
plot_waveforms_variation_with_colormap(waveforms_df, 'sc', fixed_params_sc, 'Waveform Variation with Surface Charge', tolerance=1e-4)
plot_waveforms_variation_with_colormap(waveforms_df, 'sf_drift', fixed_params_sf_drift, 'Waveform Variation with Surface Drift Factor', tolerance=1e-4)
plot_waveforms_variation_with_colormap(waveforms_df, 'pass_thickness', fixed_params_pass_thickness, 'Waveform Variation with Passivated Thickness', tolerance=1e-4)
```
